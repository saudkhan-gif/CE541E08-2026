## Section 1 — Arithmetic operators

| Symbol | Meaning | Engineering use |
|---|---|---|
| `+`, `-`, `*`, `/` | Basic arithmetic | Any formula |
| `//` | Floor division | Whole pipe sections needed |
| `%` | Modulus (remainder) | Leftover pipe length |
| `**` | Power | V², R^(2/3) |

**Same as BODMAS:** `**` → `*` `/` → `+` `-`

### ▶ Cell 1 — Hydraulic jump and pipe ordering

```python
import math

# Hydraulic jump — sequent depth
y1, V1, g = 0.4, 3.2, 9.81
Fr1 = V1 / math.sqrt(g * y1)
y2  = (y1/2) * (math.sqrt(1 + 8*Fr1**2) - 1)

print(f"Fr1 = {Fr1:.3f}, y2 = {y2:.3f} m, ratio y2/y1 = {y2/y1:.3f}")

# Pipe ordering
total_L   = 347.5
section_L = 6.0
full      = int(total_L // section_L)
remainder = total_L % section_L
print(f"Pipe: {full} × {section_L} m sections + {remainder} m off-cut")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Operator precedence in Fr formula: ** before *, then sqrt.
# Floor division (//) for ordering whole pipe sections.
# ───────────────────────────────────────────────────────────────
```

### ▶ Cell 2 — Comparison operators: design checks

```python
import math

V   = 2.8     # m/s
D   = 0.3     # m
nu  = 1.007e-6

above_min = V >= 0.6
below_max = V <= 3.0
design_ok = above_min and below_max

Re        = V * D / nu
turbulent = Re > 4000

print(f"Velocity checks: above_min={above_min}, below_max={below_max}, OK={design_ok}")
print(f"Reynolds: Re={Re:.0f}, turbulent={turbulent}")
```

### ▶ Cell 3 — Froude number for multiple flow conditions

```python
import math
g = 9.81

conditions = [
    ("Irrigation canal", 0.8, 2.1),
    ("Storm drain",      0.3, 1.8),
    ("Spillway toe",     0.15, 6.5),
    ("Reservoir inlet",  2.0, 0.4),
]

print(f"{'Location':<20} {'y':>6} {'V':>8} {'Fr':>8} {'Class'}")
print("─" * 55)
for name, y, V in conditions:
    Fr    = V / math.sqrt(g * y)
    cls   = "Subcritical" if Fr < 1 else "Supercritical"
    print(f"{name:<20} {y:>6} {V:>8} {Fr:>8.3f} {cls}")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Ask students to predict before running.
# Spillway toe: high V, shallow y → supercritical (Fr>>1)
# Reservoir inlet: deep, slow → subcritical (Fr<1)
# ───────────────────────────────────────────────────────────────
```

### ▶ Cell 4 — Logical operators: flood warning

```python
river_level   = 8.4
warning_level = 7.5
danger_level  = 9.0
rainfall_mmhr = 32.0
heavy_thresh  = 25.0
gate_open     = False

level_warn  = river_level >= warning_level
raining     = rainfall_mmhr >= heavy_thresh
danger      = river_level >= danger_level or (level_warn and gate_open)
all_clear   = not level_warn and not raining

print(f"Level warning: {level_warn}")
print(f"Heavy rain  : {raining}")
print(f"DANGER alert: {danger}")
print(f"All clear   : {all_clear}")
```

## Day 5 Assignment — Trapezoidal channel
B=4 m, m=1.5, y=1.8 m, S=0.0008, n=0.025.
Compute A, P, R, V (Manning's), Q, Fr. Check: V in 0.5–3.0 m/s? Fr<1?
*(Answers: A=12.87 m², V=1.382 m/s, Q=17.8 m³/s, Fr=0.388)*

### ▶ Assignment cell

```python
import math
B=4.0; m=1.5; y=1.8; S=0.0008; n=0.025; g=9.81
A = (B + m * y) * y
P = B + 2 * y * math.sqrt(1 + m**2)
R = A / P
T = B + 2*m*y          # top width for trapezoidal
V = (1/n) * R**(2/3) * S**(1/2)
Q = A * V
Fr = V / math.sqrt(g * A/T)   # A/T = hydraulic depth
print(f"V={V:.3f} m/s, Q={Q:.3f} m³/s, Fr={Fr:.3f}")
print(f"Velocity OK: {0.5<=V<=3.0}, Subcritical: {Fr<1}")
```
