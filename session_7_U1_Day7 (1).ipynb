## Section 1 — print() format specifications

| Spec | Meaning | Example output |
|---|---|---|
| `.4f` | 4 decimal places | `0.0345` |
| `.2e` | Scientific notation | `3.45e-02` |
| `>12` | Right-align 12 chars | `         0.0345` |
| `<12` | Left-align 12 chars | `0.0345      ` |
| `,` | Thousands separator | `1,234,567` |
| `.1%` | Percentage | `67.3%` |

### ▶ Cell 1 — Formatted engineering output table

```python
Q = 0.034521; h_f = 3.14159; Re = 98745.6

print(f"{'Parameter':<22} {'Value':>12} {'Unit'}")
print("─" * 40)
print(f"{'Discharge Q':<22} {Q:>12.4f} m³/s")
print(f"{'Discharge Q':<22} {Q*1000:>12.2f} L/s")
print(f"{'Head loss':<22} {h_f:>12.3f} m")
print(f"{'Reynolds number':<22} {Re:>12,.0f} —")
```

### ▶ Cell 2 — Interactive pipe calculator using input()

```python
import math
print('PIPE FLOW CALCULATOR — CE541E08')
diameter_mm = float(input('Pipe diameter (mm): '))
slope_ratio = float(input('Slope 1 in ___: '))
manning_n   = float(input('Manning\'s n: '))

D=diameter_mm/1000; S=1/slope_ratio
A=math.pi*(D/2)**2; P=math.pi*D; R=A/P
V=(1/manning_n)*R**(2/3)*S**0.5; Q=V*A

print(f'Velocity  : {V:.3f} m/s')
print(f'Discharge : {Q:.5f} m³/s  ({Q*1000:.2f} L/s)')

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# CLASSROOM ACTIVITY: try diameter=300, slope=500, n=0.013
# Then: diameter=450, slope=1000, n=0.013 — compare results.
# ───────────────────────────────────────────────────────────────
```

### ▶ Cell 3 — Monthly rainfall summary report

```python
monthly = {
    "Jan":4.2,"Feb":8.1,"Mar":16.5,"Apr":52.3,"May":87.6,"Jun":134.2,
    "Jul":118.4,"Aug":112.7,"Sep":95.3,"Oct":71.2,"Nov":43.8,"Dec":12.6
}
normals = {
    "Jan":5.0,"Feb":8.0,"Mar":15.0,"Apr":45.0,"May":80.0,"Jun":120.0,
    "Jul":110.0,"Aug":110.0,"Sep":90.0,"Oct":75.0,"Nov":40.0,"Dec":10.0
}
total = sum(monthly.values())
norm_t = sum(normals.values())

print(f"{'Month':<10} {'Actual':>8} {'Normal':>8} {'Anomaly':>10}")
print("─"*40)
for m in monthly:
    diff = monthly[m] - normals[m]
    print(f"{m:<10} {monthly[m]:>8.1f} {normals[m]:>8.1f} {diff:>+10.1f}")
print("─"*40)
print(f"{'ANNUAL':<10} {total:>8.1f} {norm_t:>8.1f} {total-norm_t:>+10.1f}")
print(f'Departure: {(total/norm_t-1)*100:+.1f}% of normal')
```

## Day 7 Assignment — Rainfall unit converter

Build an interactive script using `input()` that:
1. Takes: rainfall depth (mm), catchment area (km²), duration (hours)
2. Computes: depth in m, area in m², volume in m³ and Million Litres, intensity in mm/hr
3. If intensity > 64.5 mm/hr → print IMD "Heavy Rain" warning

### ▶ Assignment cell

```python
rainfall_mm   = float(input("Rainfall depth (mm): "))
area_km2      = float(input("Catchment area (km²): "))
duration_hr   = float(input("Duration (hours): "))

rainfall_m    = rainfall_mm / 1000
area_m2       = area_km2 * 1_000_000
volume_m3     = rainfall_m * area_m2
volume_ML     = volume_m3 / 1000      # 1 ML = 1000 m³
intensity_mmhr = rainfall_mm / duration_hr

print(f"Depth    : {rainfall_m:.4f} m")
print(f"Area     : {area_m2:,.0f} m²")
print(f"Volume   : {volume_m3:,.0f} m³  ({volume_ML:.3f} ML)")
print(f"Intensity: {intensity_mmhr:.1f} mm/hr")
if intensity_mmhr > 64.5:
    print("⚠ HEAVY RAIN — IMD threshold exceeded")
```
