## Section 1 — Conditions

A condition evaluates to `True` or `False`:
```python
velocity > 3.0          # True if velocity exceeds 3.0 m/s
rainfall_mm >= 64.5     # True if heavy rain threshold exceeded
material == "concrete"  # True if material matches
not data_verified       # True if data has NOT been verified
```

### ▶ Cell 1 — Writing and evaluating conditions

````python
velocity = 2.4; diameter = 300; Q_actual = 0.045; Q_design = 0.040

print("Condition evaluations:")
print(f"  velocity > 3.0      : {velocity > 3.0}")
print(f"  velocity >= 0.6     : {velocity >= 0.6}")
print(f"  diameter == 300     : {diameter == 300}")
print(f"  Q_actual > Q_design : {Q_actual > Q_design}")

vel_ok  = velocity >= 0.6 and velocity <= 3.0
cap_ok  = Q_actual >= Q_design
print()
print(f"  vel_ok (0.6–3.0)    : {vel_ok}")
print(f"  cap_ok (Q≥Qd)       : {cap_ok}")
print(f"  both OK             : {vel_ok and cap_ok}")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# Conditions return True/False — you can store them in variables.
# This is exactly what NumPy boolean arrays do for thousands
# of values simultaneously (Unit 3).
# ───────────────────────────────────────────────────────────────
````

## Section 2 — The if statement

```python
if condition:
    # runs ONLY when condition is True
    do_something()
```
The colon `:` and indentation are mandatory.

### ▶ Cell 2 — if statement: minimum pipe diameter check

````python
proposed_mm = 80   # ← try: 80, 100, 150, 300
min_mm      = 100

if proposed_mm >= min_mm:
    print(f"Diameter {proposed_mm} mm meets minimum requirement.")

print(f"Margin: {proposed_mm - min_mm} mm above minimum")

print("Check complete.")   # always runs

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# 1. Run with proposed=150 → 3 lines print
# 2. Change to 80 → only "Check complete." prints
# 3. Change to 100 → exactly at minimum, still passes (>=)
# Ask: "Which operator would make 100 fail?" Answer: >
# ───────────────────────────────────────────────────────────────
````

## Section 3 — if-else: two-way decision

```python
if condition:
    do_this()      # runs when True
else:
    do_that()      # runs when False
```
Exactly one block always runs.

### ▶ Cell 3 — if-else: subcritical or supercritical flow

````python
import math
V=3.5; y=0.4; g=9.81
Fr = V / math.sqrt(g * y)

print(f"V={V:.3f} m/s, y={y} m, Fr={Fr:.3f}")

if Fr < 1.0:
    print("SUBCRITICAL (Fr < 1) — tranquil flow")
    print("  Disturbances propagate upstream.")
else:
    print("SUPERCRITICAL (Fr > 1) — shooting flow")
    print("  Disturbances cannot propagate upstream.")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# V=3.5, y=0.4 → Fr=1.767 → supercritical
# Ask: "What max velocity keeps this subcritical?"
# Fr<1 → V < sqrt(g*y) = sqrt(9.81*0.4) = 1.98 m/s
# ───────────────────────────────────────────────────────────────
````

## Section 4 — IMD Rainfall Classification

| Category | Range (mm/day) |
|---|---|
| No rain | 0 |
| Light | 0.1–15.5 |
| Moderate | 15.6–64.4 |
| Heavy | 64.5–115.5 |
| Very Heavy | 115.6–204.4 |
| Extremely Heavy | ≥ 204.5 |

Today: two-category version. Full 6-category version in Day 12.

### ▶ Cell 4 — IMD heavy rain check

````python
rainfall_mm = 87.3   # ← try: 10, 45, 65, 120, 210
catchment_km2 = 1250.0

print(f"Daily rainfall: {rainfall_mm} mm")

if rainfall_mm >= 64.5:
    print("Classification: HEAVY or above (≥ 64.5 mm)")
    print("  Check upstream gauges — IMD Orange/Red Alert protocol.")
else:
    print("Classification: LIGHT or MODERATE (< 64.5 mm)")
    print("  Routine monitoring continues.")

volume_Mm3 = rainfall_mm/1000 * catchment_km2*1e6 / 1e6
print()
print(f"  Potential runoff on {catchment_km2} km²: {volume_Mm3:.1f} million m³")

# ─── INSTRUCTOR NOTE ───────────────────────────────────────────
# 200 mm on 1250 km² = 250 million m³ — a medium reservoir.
# This motivates why rainfall classification matters for dam safety.
# ───────────────────────────────────────────────────────────────
````

Write a script that:

### ▶ Assignment cell

````python
import math
V=1.5; D=0.3; g=9.81

if V < 0.6:
    print(f"V={V}: BELOW minimum — silting risk")
elif V >= 0.6 and V <= 3.0:
    print(f"V={V}: ACCEPTABLE ✓")
else:
    print(f"V={V}: ABOVE maximum — erosion risk")

hyd_depth = D/4
Fr = V / math.sqrt(g * hyd_depth)
if Fr < 1:
    print(f"Fr={Fr:.3f}: Subcritical")
else:
    print(f"Fr={Fr:.3f}: Supercritical")

# Test with V=0.3, 1.5, 3.8
````
